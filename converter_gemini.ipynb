```python
all_ids=[]
all_questions=list(df.iloc[0:2000]["question"])
all_answers=list(df.iloc[0:2000]["answerText"])
map_questions={str(i):"No Answer" for i in range(0,2000)}

print(len(files))
responses=[]
for i in range(1,len(files)):
    f = open(files[i],"r")
    data = f.read()
    regex_pattern_answer = r'Answer-\d*:(.*)'
    regex_pattern_id = r'Answer-(\d*):'
    ids = re.findall(regex_pattern_id,data)
    answers = re.findall(regex_pattern_answer,data)
    for id in range(len(ids)):
        if id==850:
            print("Hurrah")
        responses.append(tuple([ids[id],answers[id]]))



print(responses)
for i in responses:
    id = i[0]
    answer=i[1]
    map_questions[id]=answer
```

```python
# respones = pd.DataFrame(responses,columns=["id","answer"])
# respones = pd.DataFrame(responses,columns=["question","answer"])
answers = pd.DataFrame.from_dict(map_questions,orient='index',columns=["llm_response"])
answers["questions"]=all_questions
answers["actual_answers"]=all_answers
answers["category"]=list(df.iloc[0:2000]["category"])
answers["complexityType"]=list(df.iloc[0:2000]["complexityType"])
# Reorder the columns
answers = answers[["questions","actual_answers","llm_response","category","complexityType"]]
answers.to_csv("answers_gemini.csv",index=False)
answers
 
```
